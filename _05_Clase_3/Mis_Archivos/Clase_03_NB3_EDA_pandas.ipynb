# Intro a Análisis Exploratorio de Datos con Pandas

```python
# Cargar en una variable la URL del archivo csv de Github
url = "https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv"
```

```python
# Importar la librería Pandas
import pandas as pd
```

```python
# Usar el método read_csv de Pandas (pd) que recibe como argumento el archivo csv y retorna un dataframe
df = pd.read_csv(url)
```

```python
# Trae los primeros 5 registros del dataframe
df.head()
```

```python
# Trae los últimos 5 registros del dataframe
df.tail()
```

```python
# Muestra indo del dataframe
df.info()
```

```python
# Muestra el tipo de dato de cada campo del dataframe
df.dtypes
```

```python
# Trae estadísticas del dataframe :
#   count: Number of non-null observations.
#   mean: Average value.
#   std: Standard deviation, measuring the spread of data.
#   min: Minimum value.
#   25% (Q1): First quartile (25th percentile).
#   50% (Q2): Median (50th percentile).
#   75% (Q3): Third quartile (75th percentile).
#   max: Maximum value.
df.describe()
```

```python
# Retorna una tupla (number_of_rows, number_of_columns) del dataframe.
df.shape
```

```python
# Devuelve la cantidad de null de cada columna del dataframe.
df.isnull().sum()
```

```python
# Devuelve un objeto con todas las columnas del dataframe.
df.columns
```

```python
# Devuelve la cantidad para cada valor de determinada columna del dataframe.
df["Sex"].value_counts()
```

```python
# Realiza un gráfico con 2 librerías : seaborn y matplotlib. Les pasa el dataframe y el campo Pclass para mostrar.
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', data=df, palette='Set1')
plt.title("Distribución de la clase")
plt.xlabel("Clase")
plt.ylabel("Cantidad")
plt.show()
```
