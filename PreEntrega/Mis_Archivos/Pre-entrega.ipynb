## Script básico para carga de ventas y cálculo de total acumulado
Actividades 2 y 3

```python
# Clase para imprimir en color y negrita
class color:
   DARKCYAN = '\033[36m'
   BOLD = '\033[1m'
   END = '\033[0m'
```

```python
# Genero una lista de diccionarios (producto, precio, cantidad) con algunos datos de ejemplo
ventas = [
  {'producto': 'remera', 'precio': 9500, 'cantidad': 3},
  {'producto': 'pantalón', 'precio': 50000, 'cantidad': 1},
  {'producto': 'buzo', 'precio': 30000, 'cantidad': 2}
]
```

```python
# Calculo total de ventas con map y función lambda
total_ventas = sum(map(lambda venta: venta['precio'] * venta['cantidad'], ventas))

# Imprimo el total en negrita, color darkcyan, con el $ adelante, el punto como separador de miles en lugar de coma, y sin decimales (porque los precios tampoco tienen decimales)
print("El total de ventas es: " + color.BOLD + color.DARKCYAN + f"${'{:,.0f}'.format(total_ventas).replace(',', '.')}" + color.END)
```

```python
# Calculo total de ventas acumuladas con listas por comprensión
total_ventas = sum([venta["precio"] * venta["cantidad"] for venta in ventas])

# Imprimo el total en negrita, color darkcyan, con el $ adelante, el punto como separador de miles en lugar de coma, y sin decimales (porque los precios tampoco tienen decimales)
print("El total de ventas es: " + color.BOLD + color.DARKCYAN + f"${'{:,.0f}'.format(total_ventas).replace(',', '.')}" + color.END)
```

## Transformación de datos
Actividad 2 - Aplicar filtros y transformaciones para crear una tabla de ventas que muestre solo los productos con alto rendimiento.

```python
# Examino los primeros 2 registros
df_ventas.head(2)
```

```python
# Transformación:
# Calculo el valor de cada venta = precio * cantidad
# Agrego una nueva columna valor_venta
df_ventas["valor_venta"] = df_ventas["precio"] * df_ventas["cantidad"]
```

```python
# Visualizo que se haya generado correctamente
df_ventas.head(2)
```

```python
# Agregación:
# Agrupo valor_venta por producto
# Sumo las ventas dentro de cada grupo, generando un nuevo dataframe tabla_ventas y usando producto como índice
# tabla_ventas = (df_ventas.groupby("producto")["valor_venta"].sum())
# Si quiero que producto sea un campo normal y agregar un índice integer, puedo por pasarle el parámetro as_index=False
tabla_ventas = (df_ventas.groupby("producto", as_index=False)["valor_venta"].sum())

tabla_ventas.head()
```

```python
# Uso describe para analizar valores del dataframe tabla_ventas
tabla_ventas.describe()
```

```python
# Agrego un plot para visualizar graficamente la distribución
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(tabla_ventas["valor_venta"], bins=30, kde=True, edgecolor="black")
plt.title("Distribución de Valor de Venta con Curva de Densidad")
plt.xlabel("Valor de venta")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Calculo el percentil 75 (umbral de alto rendimiento)
umbral_ventas = tabla_ventas["valor_venta"].quantile(0.75)
umbral_ventas
```

```python
# Identifico los productos que superan el umbral
productos_top = tabla_ventas.query("valor_venta >= @umbral_ventas")
# productos_top = tabla_ventas[tabla_ventas["valor_venta"] >= umbral_ventas]
productos_top
```

## Integración de datos
Actividad 4 - Combinar los sets de datos de ventas y marketing para obtener una visión más amplia de las tendencias.

```python
df_ventas.head(2)
```

```python
df_marketing.info()
```

```python
# Uno los datasets de ventas y marketing por la columna 'producto'
ventas_marketing = pd.merge(df_ventas, df_marketing, on="producto", how="left")
```

```python
# Verifico coincidencias (muestra)
ventas_marketing.tail()
```

### Analizamos el impacto de la campaña

```python
# Calculo el costo total de los productos vendidos [cantidad] * [costo]
# Agrego una nueva columna valor_costo
ventas_marketing["valor_costo"] = ventas_marketing["cantidad"] * ventas_marketing["costo"]
ventas_marketing.head()
```

```python
# Agrego una columna con la diferencia entre lo invertido (valor_costo) y lo vendido (valor_venta)
ventas_marketing["diferencia_venta_costo"] = ventas_marketing["valor_venta"] - ventas_marketing["valor_costo"]

# Muestro solo las columnas que me interesan
ventas_marketing[['producto', 'valor_venta', 'valor_costo', 'diferencia_venta_costo']]
```

```python
# Agrupo por categoria y muestro los totales de costo, venta y diferencia de todo lo vendido
ventas_marketing.groupby("categoria")[["valor_costo", "valor_venta", "diferencia_venta_costo"]].sum()
```

```python
# Calculo la diferencia total (ganancia) entre costo y venta de todo lo vendido
ganancia_total = ventas_marketing["diferencia_venta_costo"].sum()

# Muestro el total en negrita, color darkcyan, con punto como separador de miles y con 2 decimales
print("La ganancia total es : " + color.BOLD + color.DARKCYAN + f"${'{:_.2f}'.format(ganancia_total).replace('.', ',').replace('_', '.')}" + color.END)
```
