Funcion `crear_matriz_normal()` una matriz normal, recibe número de filas y columnas (si solo pasamos número de filas, crea una matriz cuadrada)

```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp, pi, sqrt


def crear_matriz_normal(nFilas, nColumnas=None):
    if nColumnas is None:
        nColumnas = nFilas
    matriz = [[0] * nColumnas for _ in range(nFilas)]

    # Parámetros de la normal
    mu_x, mu_y = 0, 0
    sigma_x, sigma_y = 1, 1

    # Rango de coordenadas
    x_vals = np.linspace(-3, 3, nColumnas)
    y_vals = np.linspace(-3, 3, nFilas)

    for i in range(nFilas):
        for j in range(nColumnas):
            x = x_vals[j]
            y = y_vals[i]
            # Fórmula de la normal bivariada
            z = (1 / (2 * pi * sigma_x * sigma_y)) * exp(
                -(((x - mu_x) ** 2) / (2 * sigma_x**2) + ((y - mu_y) ** 2) / (2 * sigma_y**2))
            )
            matriz[i][j] = z
    return matriz
```

Función `plot_matriz()` para graficar una matriz en 3D

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt


def plot_matriz(matriz):
    Z = np.array(matriz)  # conversión para graficar
    n, m = Z.shape
    x_vals = np.linspace(-3, 3, m)
    y_vals = np.linspace(-3, 3, n)
    X, Y = np.meshgrid(x_vals, y_vals)
    fig = plt.figure(figsize=(7, 7))
    ax = fig.add_subplot(111, projection="3d")
    ax.plot_surface(X, Y, Z, edgecolor="k")  # simple y claro
    ax.set_title("Plot")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Altura")
    plt.show()
```
