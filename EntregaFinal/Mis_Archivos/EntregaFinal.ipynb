## Script básico para carga de ventas y cálculo de total acumulado
Actividades 2 y 3

```python
# Clase para imprimir en color y negrita
class color:
   DARKCYAN = '\033[36m'
   BOLD = '\033[1m'
   END = '\033[0m'
```

```python
# Genero una lista de diccionarios (producto, precio, cantidad) con algunos datos de ejemplo
ventas = [
  {'producto': 'remera', 'precio': 9500, 'cantidad': 3},
  {'producto': 'pantalón', 'precio': 50000, 'cantidad': 1},
  {'producto': 'buzo', 'precio': 30000, 'cantidad': 2}
]
```

```python
# Calculo total de ventas con map y función lambda
total_ventas = sum(map(lambda venta: venta['precio'] * venta['cantidad'], ventas))

# Imprimo el total en negrita, color darkcyan, con el $ adelante, el punto como separador de miles en lugar de coma, y sin decimales (porque los precios tampoco tienen decimales)
print("El total de ventas es: " + color.BOLD + color.DARKCYAN + f"${'{:,.0f}'.format(total_ventas).replace(',', '.')}" + color.END)
```

```python
# Calculo total de ventas acumuladas con listas por comprensión
total_ventas = sum([venta["precio"] * venta["cantidad"] for venta in ventas])

# Imprimo el total en negrita, color darkcyan, con el $ adelante, el punto como separador de miles en lugar de coma, y sin decimales (porque los precios tampoco tienen decimales)
print("El total de ventas es: " + color.BOLD + color.DARKCYAN + f"${'{:,.0f}'.format(total_ventas).replace(',', '.')}" + color.END)
```

### Unión de los dataframes de ventas y marketing

```python
# Veo que en ambos datasets está la columna 'producto'. Los uno entonces por esa columna
ventas_marketing = pd.merge(df_ventas, df_marketing, on="producto", how="left")
print("Cantidad de registros: ", len(ventas_marketing))
```

```python
# Veo algunos registros luego del merge (ordenados por producto, id_venta)
ventas_marketing[["id_venta","producto","canal","cantidad","precio","valor_venta","costo","fecha_venta","fecha_inicio","fecha_fin"]].sort_values(by=["producto", "id_venta"], ascending=True).head(6)
```

```python
# Veo algunos registros luego del merge, pero ordenados por fecha_venta
ventas_marketing[["id_venta","producto", "valor_venta", "canal", "costo", "fecha_venta", "fecha_inicio","fecha_fin"]].sort_values(by=["valor_venta"], ascending=False).head(6)
```

**Luego del merge se encuentra:**

**Aparecen ventas triplicadas por el hecho de que hay 3 campañas de marketing (RSS, TV, Email) por cada producto.**
En la vista anterior se ven los primeros 3 registros con id_venta 2063, pero encima en ese caso el producto fue vendido el 5/3/24, fuera de los 3 períodos de marketing para ese producto.
Y en el caso de los últimos 3 registros con id_venta 1109, dos de ellos fuera, y uno dentro de los 3 períodos de marketing.
Por ello, voy a generar dos nuevos dataframes, uno con las ventas dentro del período de marketing, y otro con las ventas fuera del período.

### Ventas dentro de la campaña (opción con map, y opción con filtro, agrupación y agregación)

```python
# Agrego con map una columna "en_campaña" SI o NO en ventas_marketing
ventas_marketing['en_campaña'] = (
    (ventas_marketing['fecha_venta'] >= ventas_marketing['fecha_inicio']) &
    (ventas_marketing['fecha_venta'] <= ventas_marketing['fecha_fin'])
).map({True: "SI", False: "NO"})
```

```python
# Veo cómo quedó con algunas filas de ejemplo
ventas_marketing[["id_venta","producto","fecha_venta","fecha_inicio","fecha_fin","en_campaña"]].sample(10)
```

```python
# Veo el total de cada uno
ventas_marketing["en_campaña"].value_counts()
```

```python
# Filtro solo aquellos productos que se hayan vendido dentro de una campaña de marketing
# ventas_con_marketing = ventas_marketing.query("en_campaña == 'SI'")
ventas_con_marketing = ventas_marketing.query("fecha_venta >= fecha_inicio & fecha_venta <= fecha_fin")
print("Cantidad de registros dentro de la campaña: ", len(ventas_con_marketing))
```

```python
# Visualizo algunos de los registros de ventas durante la campaña de marketing
ventas_con_marketing[["id_venta","producto", "valor_venta", "canal", "costo", "fecha_venta", "fecha_inicio","fecha_fin"]].sort_values(by=["producto", "id_venta"], ascending=True).head(6)
```

```python
# Como hay duplicados, listo solo esos
ventas_con_marketing_duplicados = ventas_con_marketing[ventas_con_marketing.duplicated(subset="id_venta",keep=False)]
ventas_con_marketing_duplicados[["id_venta","producto", "valor_venta", "canal", "costo", "fecha_venta", "fecha_inicio","fecha_fin"]].sort_values(by=["producto", "id_venta"], ascending=True).head(6)
```

Los registros con id_venta duplicados nos impiden agrupar directamente por producto, debemos antes agrupar por [id_venta, producto] quedarnos con el primer valor_venta y hacer la suma de los costos de marketing.
Luego sí agrupamos por producto, para saber el monto total de venta para cada uno.

```python
# Agrupo primero por id_venta y producto
# De esa partición, me quedo con la primer venta y la suma de los costos de marketing
ventas_con_marketing_agg1 = ventas_con_marketing.groupby(["id_venta","producto"], as_index=False).agg(
    valor_venta=("valor_venta", "first"),
    costo_agg=("costo", "sum")
)

print("Cantidad de registros con id_venta únicos: ", len(ventas_con_marketing_agg1))
```

```python
# Verifico la consistencia de los montos, ventas y costos de marketing
ventas_con_marketing_agg1[["id_venta","producto", "valor_venta", "costo_agg"]].sort_values(by=["producto", "id_venta"], ascending=True).head(6)
```

```python
# En la segunda agregacion, agrupo solo por producto
ventas_con_marketing_agg2 = ventas_con_marketing_agg1.groupby(["producto"], as_index=False).agg(
    ventas_producto_cm=("valor_venta", "sum"),
    costo_marketing=("costo_agg", "sum")
)

# Agrego otra columna con la ganancia
ventas_con_marketing_agg2["ganacia_producto_cm"]= ventas_con_marketing_agg2["ventas_producto_cm"]-ventas_con_marketing_agg2["costo_marketing"]

# Finalmente, ordeno y guardo en el dataframe final
ventas_con_marketing_final = ventas_con_marketing_agg2.sort_values(by=["ganacia_producto_cm"], ascending=False)
```

```python
# Dataframe resultante
ventas_con_marketing_final.head(5)
```

### Ventas fuera de la campaña de marketing

```python
# Filtramos solo aquellos productos que se hayan vendido fuera de la campaña
# ventas_sin_marketing = ventas_marketing.query("en_campaña == 'NO'")
ventas_sin_marketing = ventas_marketing.query("(fecha_venta < fecha_inicio) | (fecha_venta > fecha_fin)")
print("Cantidad de registros fuera de la campaña: ", len(ventas_sin_marketing))
```

```python
# Visualizo algunos de los registros de ventas fuera de la campaña de marketing
ventas_sin_marketing[["id_venta","producto", "valor_venta", "fecha_venta", "fecha_inicio","fecha_fin"]].sort_values(by=["id_venta","producto"], ascending=True).head(6)
```

```python
# Acá también hay id_venta duplicados, así que procedo igual que antes, agrupando primero por id_venta y producto
# y luego solo por producto
ventas_sin_marketing_agg1 = ventas_sin_marketing.groupby(["id_venta", "producto"], as_index=False).agg(
    valor_venta=("valor_venta", "first")
)
print(len(ventas_sin_marketing_agg1))

ventas_sin_marketing_agg2 = ventas_sin_marketing_agg1.groupby(["producto"], as_index=False).agg(
    ventas_producto_sm=("valor_venta", "sum")
)

# Ordeno y guardo en el dataframe final
ventas_sin_marketing_final = ventas_sin_marketing_agg2[["producto", "ventas_producto_sm"]].sort_values(by=["ventas_producto_sm"], ascending=False)
```

```python
ventas_sin_marketing_final.head(5)
```

```python
# Nightingale Rose
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def nightingale_rose(df, col_categoria="mes", col_valor="cantidades_mes",
                     title="Nightingale Rose Diagram"):
    """
    df: DataFrame ordenado con dos columnas
    col_categoria: nombre de la columna de categorías
    col_valor: nombre de la columna numérica
    """

    categorias = df[col_categoria].tolist()
    valores = df[col_valor].tolist()

    N = len(df)                           # número de filas del df
    angles = np.linspace(0, 2*np.pi, N, endpoint=False)

    # Paleta automática (tab20 soporta hasta 20 colores, tab20b + tab20c permiten más)
    if N <= 15:
        colors = plt.cm.tab20(np.linspace(0, 1, N))
    else:
        # Para más de 20 categorías, usamos una paleta continua
        colors = plt.cm.viridis(np.linspace(0, 1, N))

    plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, polar=True)

    # Barras
    ax.bar(
        angles,
        valores,
        width=2 * np.pi / N,
        bottom=0,
        color=colors,
        edgecolor="white",
        linewidth=1
    )

    # Etiquetas
    for angle, value, label in zip(angles, valores, categorias):
        ax.text(
            angle,
            value + (max(valores) * 0.05),
            f"{label}\n{value}",
            ha="center",
            va="center",
            fontsize=12,
            color="black",
            fontweight="bold"
        )

    # Limpieza estética
    ax.set_yticklabels([])
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)

    plt.title(title, fontsize=16, pad=40)
    plt.show()
```
